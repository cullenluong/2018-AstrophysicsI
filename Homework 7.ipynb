Impact parameter and sepeartion, and inclination angle between planet and star forms a right triangle where

leg = $br_{*} $
hypotenus = $a $
The inclination angle is related by the definition of cosine in the right triangle.

$\cos i =  \frac{br_{*}}{a}$

To find the transit duration  we need to know velocity and distance. 
The velocity is simply the orbital velocity

$v = \frac{2\pi a}{\tau} $

The distance is 2 times the other leg of the above right triangle

$d = 2r_{*}\sqrt{1-b^2} $

Thus we get

$t_{\mathrm{duration}} = v/d = r_{*} \tau \sqrt{1-b^2}/a $
